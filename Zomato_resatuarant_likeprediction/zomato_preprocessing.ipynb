```python
import pandas as pd
import json
from IPython.display import display

# Step 1: Load JSON File
def load_json_file(file_path):
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            data = json.load(file)
        return data
    except Exception as e:
        print(f"Error loading JSON file: {e}")
        return []

# Step 2: Extract and Transform
def extract_restaurant_data(data):
    restaurant_details = []
    for entry in data:
        if "restaurants" in entry:
            for restaurant_entry in entry["restaurants"]:
                restaurant = restaurant_entry.get("restaurant", {})
                if restaurant:
                    # Top-level fields
                    details = {
                        "Name": restaurant.get("name"),
                        "Cuisines": restaurant.get("cuisines"),
                        "Average_Cost_For_Two": restaurant.get("average_cost_for_two"),
                        "Price_Range": restaurant.get("price_range"),
                        "Has_Online_Delivery": restaurant.get("has_online_delivery"),
                        "Is_Delivering_Now": restaurant.get("is_delivering_now"),
                        "Menu_URL": restaurant.get("menu_url"),
                        "Booking_URL": restaurant.get("book_url"),
                        "Featured_Image": restaurant.get("featured_image"),
                        "Latitude": restaurant.get("location", {}).get("latitude"),
                        "Longitude": restaurant.get("location", {}).get("longitude"),
                        "Address": restaurant.get("location", {}).get("address"),
                        "City": restaurant.get("location", {}).get("city"),
                        "Aggregate_Rating": restaurant.get("user_rating", {}).get("aggregate_rating"),
                        "Rating_Text": restaurant.get("user_rating", {}).get("rating_text"),
                        "Votes": restaurant.get("user_rating", {}).get("votes"),
                        "Res_ID": restaurant.get("R", {}).get("res_id"),
                    }
                    # Append the details
                    restaurant_details.append(details)
    return pd.DataFrame(restaurant_details)

# Step 3: Process Data and Filter Errors
def process_data(file_path):
    data = load_json_file(file_path)
    # Remove entries with API errors
    valid_data = [entry for entry in data if "message" not in entry]
    return extract_restaurant_data(valid_data)

# Step 4: Save to Excel
def save_to_excel(dataframe, output_file):
    try:
        dataframe.to_excel(output_file, index=False, engine="openpyxl")
        print(f"Data successfully saved to {output_file}")
    except Exception as e:
        print(f"Error saving to Excel: {e}")

# Main Execution
if __name__ == "__main__":
    # Prompt user for JSON file location
    file_path = input("Enter the full path to your JSON file: ").strip()
    
    # Process the data
    restaurant_df = process_data(file_path)
    
    # Display the resulting DataFrame in tabular format
    display(restaurant_df)
    
    # Save the DataFrame to an Excel file
    output_file = "file5.xlsx"
    save_to_excel(restaurant_df, output_file)

```
