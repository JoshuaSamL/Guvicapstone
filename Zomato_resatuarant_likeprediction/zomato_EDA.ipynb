```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the preprocessed Excel file
file_path = "J:\Studies\GUVI\projects\capstone\Zomato\zomato.xlsx"  # Replace with the actual file path
df = pd.read_excel(file_path)
```

```python
# Display the first few rows of the dataset
print("Dataset Preview:")
print(df.head())
```

```python
# Plot distributions
plt.figure(figsize=(10, 6))
sns.histplot(df['Average_Cost_For_Two'], kde=True, bins=30)
plt.title("Distribution of Average Cost for Two")
plt.xlabel("Average Cost for Two")
plt.ylabel("Frequency")
plt.show()
```

```python
# Filter numeric columns for correlation calculation
numeric_df = df.select_dtypes(include=["float64", "int64"])

# Check if numeric columns exist
if not numeric_df.empty:
    plt.figure(figsize=(10, 8))
    sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
    plt.title("Correlation Matrix")
    plt.show()
else:
    print("No numeric columns found for correlation analysis.")
```

```python
# Countplot for price ranges
plt.figure(figsize=(8, 6))
sns.countplot(x="Price_Range", data=df, palette="viridis")
plt.title("Count of Restaurants by Price Range")
plt.xlabel("Price Range")
plt.ylabel("Count")
plt.show()
```
